```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, learning_rate=0.01, iterations=10000):
    # Ensure x and y are numpy arrays of floats
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)
    
    m_curr = b_curr = 0
    n = len(x)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
    return m_curr, b_curr

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 15])

# Perform gradient descent
m, b = gradient_descent(x, y)

# Plotting the data points and the fitted line
plt.scatter(x, y, color='red', marker='+', linewidth=5)
plt.plot(x, m * x + b, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.show()

# Print the final parameters
print(f"Final parameters: m = {m}, b = {b}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, learning_rate=0.01, iterations=10000):
    # Ensure x and y are numpy arrays of floats
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)
    
    m_curr = b_curr = 0
    n = len(x)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
    return m_curr, b_curr

def calculate_mse(x, y, m, b):
    y_pred = m * x + b
    mse = np.mean((y - y_pred) ** 2)
    return mse

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 15])

# Different configurations for gradient descent
configs = [
    {'learning_rate': 0.01, 'iterations': 10000},
    {'learning_rate': 0.005, 'iterations': 10000},
    {'learning_rate': 0.01, 'iterations': 20000},
    {'learning_rate': 0.001, 'iterations': 10000},
    {'learning_rate': 0.01, 'iterations': 5000}
]

results = []

# Run gradient descent for each configuration
for config in configs:
    m, b = gradient_descent(x, y, learning_rate=config['learning_rate'], iterations=config['iterations'])
    mse = calculate_mse(x, y, m, b)
    results.append((m, b, mse))

# Sort results by MSE
results.sort(key=lambda x: x[2])

# Get the top 3 best fit lines
top_3 = results[:3]

# Plot the data points and the top 3 fitted lines
plt.scatter(x, y, color='red', marker='+', linewidth=5)
colors = ['green', 'blue', 'orange']

for i, (m, b, mse) in enumerate(top_3):
    plt.plot(x, m * x + b, color=colors[i], label=f'Line {i+1}: y={m:.2f}x + {b:.2f} (MSE={mse:.2f})')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Top 3 Linear Regression Fits with Gradient Descent')
plt.legend()
plt.show()

# Print the top 3 parameters
for i, (m, b, mse) in enumerate(top_3):
    print(f"Line {i+1}: m = {m}, b = {b}, MSE = {mse}")
```
